```python
#5.Outliers
animeova=Report.loc[(Report["type"] == "OVA"), ["name", "episodes", "rating"]]
animeova
```

```python
animeova.describe()
```

```python
sns.boxplot(data = animeova, x='rating',color='orange')
```

```python
#Extracting Outliers
# Calculate the IQR
q3 = animeova['rating'].quantile(0.75)
q1 = animeova['rating'].quantile(0.25)
IQR = q3 - q1
IQR
```

```python
# Setting the upperbound and lowerbound
upperrange = q3 + 1.5*IQR
bottomrange = q1 - 1.5*IQR
print (upperrange, bottomrange)
```

```python
animeova_outliers_df = animeova[animeova['rating'] > upperrange]
animeova_outliers_df.shape
```

```python
animeova_df2 = animeova[animeova['rating'] < upperrange]
animeova_df2
```

```python
sns.boxplot(data = animeova_df2, x='rating')
```

```python
#6. Counting each type of Anime and plotting a pie plot
types_rep = Report.type.value_counts()
types_rep 
```

### Asking & Answering questions about the data

Questions:
1. Which are the top 10 highest rated animes which have been watched by atleast 20,000 members?
2. The most high rated animes belong to which type and their percentages?
3. Which Studio produces most number of the anime from Manga source?
4. Comparison of Total number of people who watch, Total number of people scored and total number of people who favorited the animes

```python
Report_new = pd.read_csv('C:/Users/dhris/data/dataanime.csv')
Report_new
```

```python
Merged_new = pd.merge(Report, Report_new, how='inner', left_on='name', right_on='Title')
Merged_new
```

```python
Merged_new.drop(Merged_new.columns[[0,2,3,4,5,6,7]], axis = 1, inplace = True)
Merged_new.head(5)
```

#### Q1.The top 10 highest rated animes which have been watched by atleast 20,000 members

```python
#Drop rows with Members<20000 
Merged_new.drop(Merged_new[Merged_new.Members<20000].index, inplace= True)
```

```python
Merged_new.sort_values(by='Score',ascending=False).head(10)
```

```python
print(Merged_new.loc[:, ['name', 'Score']])
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x='Score', y='name',  data=Merged_new.head(10))
plt.style.use('_classic_test_patch')
plt.title("The top 10 highest rated anime")
```

#### Q2.  The most high rated animes belong to which type and its percentage?

```python
#Applying value_counts, sorting in ascending order and displaying top 10
genre_sort = Merged_new.Genres.value_counts().sort_values(ascending=False).head(15)
genre_sort
```

```python
genre_sort.plot.pie(label = ' ',autopct='%1.1f%%', title = 'High-rated Anime Genres')
plt.style.use('seaborn-colorblind')
plt.show()
```

#### Q3. Which Studio produces most number of the anime from Manga source?

```python
anime_studio=Merged_new.loc[:, ['name', 'Sources','Studios']]
anime_studio
```

```python
anime_studio_manga=Merged_new.loc[(Merged_new["Sources"] == "Manga"), ["name", "Studios"]]
anime_studio_manga
```

```python
n = 10
anime_studio_manga['Studios'].value_counts().nlargest(10)
```

```python
plt.figure(figsize = (15,15))
plt.plot(anime_studio_manga['Studios'].value_counts().nlargest(10), color='red',linewidth = 3,linestyle= '--')
plt.style.use('seaborn-dark-palette')
plt.xlabel('Studios')
plt.ylabel('Count')
plt.title('Largest no.of Studios producing anime from Manga source')
```
