```python
#2.Average no.of Members who watch anime
anime_df = Report[['name', 'genre', 'members']]
anime_df
```

```python
print("The Mean number of people watching anime is :\n", anime_df.select_dtypes(include='int64').mean())
```

```python
#3.Filtering Rating greater than 8
Report_df = Report[Report['rating'] > 8]
Report_df.head(50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline 
```

```python
#Plotting Bar Graph
Report_df.head(50).plot.bar(x='rating', y='members',color='olive',figsize=(15,5),title='Rating greater than 8');
```

```python
#4.Sorting by anime which is of the type TV
Report_sort = Report_df[Report_df['type']=='TV']
Report_sort.head(100)
```

```python
#Plotting Histogram
sns.histplot(data=Report_sort, x="rating", binwidth=0.2, color="violet", edgecolor="black").set(title='Ratings of Anime Type TV')
```

```python
#5.Outliers
animeova=Report.loc[(Report["type"] == "OVA"), ["name", "episodes", "rating"]]
animeova
```

```python
animeova.describe()
```

```python
sns.boxplot(data = animeova, x='rating',color='orange')
```

```python
#Extracting Outliers
# Calculate the IQR
q3 = animeova['rating'].quantile(0.75)
q1 = animeova['rating'].quantile(0.25)
IQR = q3 - q1
IQR
```

```python
# Setting the upperbound and lowerbound
upperrange = q3 + 1.5*IQR
bottomrange = q1 - 1.5*IQR
print (upperrange, bottomrange)
```

```python
animeova_outliers_df = animeova[animeova['rating'] > upperrange]
animeova_outliers_df.shape
```

```python
animeova_df2 = animeova[animeova['rating'] < upperrange]
animeova_df2
```

```python
sns.boxplot(data = animeova_df2, x='rating')
```
